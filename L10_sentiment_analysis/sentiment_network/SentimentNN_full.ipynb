# Project_4. Build a Neural Network

```python
import sys
import time
import numpy as np

class SentimentNetwork_1:
    def __init__(self, reviews, labels, hidden_nodes = 10, learning_rate = 0.1):
        
        # set our random number generator
        np.random.seed(1)
        
        self.pre_process_data(reviews, labels)
        self.init_network(len(self.review_vocab), hidden_nodes, 1, learning_rate)
        
    def pre_process_data(self, reviews, labels):
        review_vocab = set()
        for review in reviews:
            for word in review.split(' '):
                review_vocab.add(word)
        self.review_vocab = list(review_vocab)
        
        label_vocab = set()
        for label in labels:
            for word in label.split(' '):
                label_vocab.add(word)
        self.label_vocab = list(label_vocab)
        
        self.word2index = {}
        for idx, word in enumerate(review_vocab):
            self.word2index[word] = idx
        
        self.label2index = {}
        for idx, label in enumerate(label_vocab):
            self.label2index[label] = idx
        
        
    def init_network(self, input_nodes, hidden_nodes, output_nodes, learning_rate):
        # Set number of nodes in input, hidden, and output layers.
        self.input_nodes = input_nodes
        self.hidden_nodes = hidden_nodes
        self.output_nodes = output_nodes
        
        # Initialize the weights
        self.weights_0_1 = np.zeros((self.input_nodes, self.hidden_nodes))
        
        self.weights_1_2 = np.random.normal(0.0, self.output_nodes**-0.5, (self.hidden_nodes, self.output_nodes))
        
        self.lr = learning_rate
        
        self.layer_0 = np.zeros((1, input_nodes))
        
    def update_input_data(self, review):
        self.layer_0 *= 0 
        
        for word in review.split(' '):
            if (word in self.word2index.keys()):
                self.layer_0[0][self.word2index[word]] += 1
            
    def get_target_for_label(self, label):
        if label == 'POSITIVE':
            return 1
        else:
            return 0
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_output_2_derivative(self, output):
        return output * (1 - output)
    
    def train(self, training_reviews, training_labels):
        assert(len(training_reviews) == len(training_labels))
        
        correct_so_far = 0
        
        start = time.time()
        
        for i in range(len(training_reviews)):
            review = training_reviews[i]
            label = training_labels[i]
            
            ### Implement the forward pass here ###
            ## Forward pass
            
            # Input layer
            self.update_input_data(review)
            
            # Hidden layer
            hidden_layer = np.dot(self.layer_0, self.weights_0_1)
                        
            # Output layer
            output = self.sigmoid(np.dot(hidden_layer, self.weights_1_2))
            
            ### Implement the backward pass here ###
            ## Backward pass
            
            # Output Error
            output_error = self.get_target_for_label(label) - output
            output_error_d = output_error*self.sigmoid_output_2_derivative(output)
            
            # Backpropagated Error
            hidden_error = np.dot(output_error_d, self.weights_1_2.T)
            hidden_error_d = hidden_error
            
            # Updates the weights
            self.weights_1_2 += self.lr*np.dot(hidden_layer.T, output_error_d)
            self.weights_0_1 += self.lr*np.dot(self.layer_0.T, hidden_error_d)
            
            if np.abs(output_error) < 0.5:
                correct_so_far += 1
            
            review_per_second = i / float(time.time() - start)

            sys.stdout.write('\rProcess:' + str(100*i/float(len(training_reviews)))[:4] + \
                             '% Speed(reviews/sec):' + str(review_per_second)[:5] + \
                            '% #Correct:' + str(correct_so_far) + ' #Trained:' + str(i+1) \
                            + ' Training Accuracy:' + str(correct_so_far*100/float(i+1))[:4])
            if i % 2500 == 0:
                print(" ")
                
    def test(self, testing_reviews, testing_labels):
        correct = 0
        
        start = time.time()
        
        for i in range(len(testing_reviews)):
            pred = self.run(testing_reviews[i])
            if (pred == testing_labels[i]):
                correct += 1
                
            review_per_second = i / float(time.time() - start)
            
            sys.stdout.write('\rProcess:' + str(100*i/float(len(testing_reviews)))[:4] + \
                             '% Speed(reviews/sec):' + str(review_per_second)[:5] + \
                            '% #Correct:' + str(correct) + ' #Tested:' + str(i+1) \
                            + ' Testing Accuracy:' + str(correct*100/float(i+1))[:4])
            
        
    def run(self, review):
        # Input layer
        self.update_input_data(review.lower())
        
        # Hidden layer
        hidden_layer = np.dot(self.layer_0, self.weights_0_1)
        
        # Output layer
        output = self.sigmoid(np.dot(hidden_layer, self.weights_1_2))
        
        if output[0] > 0.5:
            return 'POSITIVE'
        else:
            return 'NEGATIVE'
                
```

```python
import sys
import time
import numpy as np

class SentimentNetwork_3:
    def __init__(self, reviews, labels, hidden_nodes = 10, learning_rate = 0.1):
        
        # set our random number generator
        np.random.seed(1)
        
        self.pre_process_data(reviews, labels)
        self.init_network(len(self.review_vocab), hidden_nodes, 1, learning_rate)
        
    def pre_process_data(self, reviews, labels):
        review_vocab = set()
        for review in reviews:
            for word in review.split(' '):
                review_vocab.add(word)
        self.review_vocab = list(review_vocab)
        
        label_vocab = set()
        for label in labels:
            for word in label.split(' '):
                label_vocab.add(word)
        self.label_vocab = list(label_vocab)
        
        self.word2index = {}
        for idx, word in enumerate(review_vocab):
            self.word2index[word] = idx
        
        self.label2index = {}
        for idx, label in enumerate(label_vocab):
            self.label2index[label] = idx
        
        
    def init_network(self, input_nodes, hidden_nodes, output_nodes, learning_rate):
        # Set number of nodes in input, hidden, and output layers.
        self.input_nodes = input_nodes
        self.hidden_nodes = hidden_nodes
        self.output_nodes = output_nodes
        
        # Initialize the weights
        self.weights_0_1 = np.zeros((self.input_nodes, self.hidden_nodes))
        
        self.weights_1_2 = np.random.normal(0.0, self.output_nodes**-0.5, (self.hidden_nodes, self.output_nodes))
        
        self.lr = learning_rate
        
        self.layer_0 = np.zeros((1, input_nodes))
        self.hidden_layer = np.zeros((1, self.hidden_nodes))
        
    def update_input_data(self, review):
        self.layer_0 *= 0 
        #self.layer_0_idx = []
        
        for word in review.split(' '):
            if (word in self.word2index.keys()):
                self.layer_0[0][self.word2index[word]] = 1  ### Modify to reduce noise in Input data
                #self.layer_0_idx.append(self.word2index[word])
                
    def get_target_for_label(self, label):
        if label == 'POSITIVE':
            return 1
        else:
            return 0
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_output_2_derivative(self, output):
        return output * (1 - output)
    
    def train(self, training_reviews_raw, training_labels):
        training_reviews = list()
        for review in training_reviews_raw:
            indices = set()
            for word in review.split(' '):
                indices.add(self.word2index[word])
            training_reviews.append(list(indices))
        
        assert(len(training_reviews) == len(training_labels))
        
        correct_so_far = 0
        
        start = time.time()
        
        for i in range(len(training_reviews)):
            review = training_reviews[i]
            label = training_labels[i]
            
            ### Implement the forward pass here ###
            ## Forward pass
            
            # Input layer
               
            # Hidden layer
            self.hidden_layer *= 0
            for index in review:
                self.hidden_layer += self.weights_0_1[index]
                        
            # Output layer
            output = self.sigmoid(np.dot(self.hidden_layer, self.weights_1_2))
            
            ### Implement the backward pass here ###
            ## Backward pass
            
            # Output Error
            output_error = self.get_target_for_label(label) - output
            output_error_d = output_error*self.sigmoid_output_2_derivative(output)
            
            # Backpropagated Error
            hidden_error = np.dot(output_error_d, self.weights_1_2.T)
            hidden_error_d = hidden_error
            
            # Updates the weights
            self.weights_1_2 += self.lr*np.dot(self.hidden_layer.T, output_error_d)
            
            for idx in review:
                self.weights_0_1[idx] += self.lr*hidden_error_d[0]
            
            if np.abs(output_error) < 0.5:
                correct_so_far += 1
            
            review_per_second = i / float(time.time() - start)

            sys.stdout.write('\rProcess:' + str(100*i/float(len(training_reviews)))[:4] + \
                             '% Speed(reviews/sec):' + str(review_per_second)[:5] + \
                            '% #Correct:' + str(correct_so_far) + ' #Trained:' + str(i+1) \
                            + ' Training Accuracy:' + str(correct_so_far*100/float(i+1))[:4])
            if i % 2500 == 0:
                print(" ")
                
    def test(self, testing_reviews, testing_labels):
        correct = 0
        
        start = time.time()
        
        for i in range(len(testing_reviews)):
            pred = self.run(testing_reviews[i])
            if (pred == testing_labels[i]):
                correct += 1
                
            review_per_second = i / float(time.time() - start)
            
            sys.stdout.write('\rProcess:' + str(100*i/float(len(testing_reviews)))[:4] + \
                             '% Speed(reviews/sec):' + str(review_per_second)[:5] + \
                            '% #Correct:' + str(correct) + ' #Tested:' + str(i+1) \
                            + ' Testing Accuracy:' + str(correct*100/float(i+1))[:4])
            
        
    def run(self, review):
        # Input layer
              
        # Hidden layer
        self.hidden_layer *= 0
        unique_indices = set()
        for word in review.lower().split(' '):
            if word in self.word2index.keys():
                unique_indices.add(self.word2index[word])
        for idx in unique_indices:
            self.hidden_layer += self.weights_0_1[idx]
        
        # Output layer
        output = self.sigmoid(np.dot(self.hidden_layer, self.weights_1_2))
        
        if output[0] > 0.5:
            return 'POSITIVE'
        else:
            return 'NEGATIVE'
                
```

# Reducing Noise by Strategically Reducing the Vocabulary

```python
import sys
import time
import numpy as np

class SentimentNetwork_4:
    def __init__(self, reviews, labels, polarity_cutoff = 0.1, min_count=10, hidden_nodes = 10, learning_rate = 0.1):
        
        # set our random number generator
        np.random.seed(1)
        
        self.pre_process_data(reviews, polarity_cutoff, min_count)
        self.init_network(len(self.review_vocab), hidden_nodes, 1, learning_rate)
        
    def pre_process_data(self, reviews, polarity_cutoff, min_count):
        positive_counts = Counter()
        negative_counts = Counter()
        total_counts = Counter()
        
        for i in range(len(reviews)):
            if labels[i] == 'POSITIVE':
                for word in reviews[i].split(' '):
                    positive_counts[word] += 1
                    total_counts[word] += 1
            else:
                for word in reviews[i].split(' '):
                    negative_counts[word] += 1
                    total_counts[word] += 1
        
        pos_neg_ratios = Counter()
        
        for term, cnt in list(total_counts.most_common()):
            if cnt >= 50 :
                pos_neg_ratio = positive_counts[term]/float(negative_counts[word] + 1)
                pos_neg_ratios[term] = pos_neg_ratio
                
        for word, ratio in pos_neg_ratios.most_common():
            if ratio > 1:
                pos_neg_ratios[word] = np.log(ratio)
            else:
                pos_neg_ratios[word] = -np.log((1/(ratio + 0.01)))
        
        review_vocab = set()
        for review in reviews:
            for word in review.split(' '):
                if (total_counts[word] > min_count):
                    if (word in pos_neg_ratios.keys()):
                        if ((pos_neg_ratios[word] >= polarity_cutoff) or (pos_neg_ratios[word] <= -polarity_cutoff)):
                            review_vocab.add(word)
                    else:
                        review_vocab.add(word)
        self.review_vocab = list(review_vocab)
        
        self.word2index = {}
        for idx, word in enumerate(review_vocab):
            self.word2index[word] = idx

    def init_network(self, input_nodes, hidden_nodes, output_nodes, learning_rate):
        # Set number of nodes in input, hidden, and output layers.
        self.input_nodes = input_nodes
        self.hidden_nodes = hidden_nodes
        self.output_nodes = output_nodes
        
        # Initialize the weights
        self.weights_0_1 = np.zeros((self.input_nodes, self.hidden_nodes))
        
        self.weights_1_2 = np.random.normal(0.0, self.output_nodes**-0.5, (self.hidden_nodes, self.output_nodes))
        
        self.lr = learning_rate
        
        self.layer_0 = np.zeros((1, input_nodes))
        self.hidden_layer = np.zeros((1, self.hidden_nodes))
        
    def update_input_data(self, review):
        self.layer_0 *= 0 
        #self.layer_0_idx = []
        
        for word in review.split(' '):
            if (word in self.word2index.keys()):
                self.layer_0[0][self.word2index[word]] = 1  ### Modify to reduce noise in Input data
                #self.layer_0_idx.append(self.word2index[word])
                
    def get_target_for_label(self, label):
        if label == 'POSITIVE':
            return 1
        else:
            return 0
        
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_output_2_derivative(self, output):
        return output * (1 - output)
    
    def train(self, training_reviews_raw, training_labels):
        training_reviews = list()
        for review in training_reviews_raw:
            indices = set()
            for word in review.split(' '):
                if (word in self.word2index.keys()):
                    indices.add(self.word2index[word])
            training_reviews.append(list(indices))
        
        assert(len(training_reviews) == len(training_labels))
        
        correct_so_far = 0
        
        start = time.time()
        
        for i in range(len(training_reviews)):
            review = training_reviews[i]
            label = training_labels[i]
            
            ### Implement the forward pass here ###
            ## Forward pass
            
            # Input layer
               
            # Hidden layer
            self.hidden_layer *= 0
            for index in review:
                self.hidden_layer += self.weights_0_1[index]
                        
            # Output layer
            output = self.sigmoid(np.dot(self.hidden_layer, self.weights_1_2))
            
            ### Implement the backward pass here ###
            ## Backward pass
            
            # Output Error
            output_error = self.get_target_for_label(label) - output
            output_error_d = output_error*self.sigmoid_output_2_derivative(output)
            
            # Backpropagated Error
            hidden_error = np.dot(output_error_d, self.weights_1_2.T)
            hidden_error_d = hidden_error
            
            # Updates the weights
            self.weights_1_2 += self.lr*np.dot(self.hidden_layer.T, output_error_d)
            
            for idx in review:
                self.weights_0_1[idx] += self.lr*hidden_error_d[0]
            
            if np.abs(output_error) < 0.5:
                correct_so_far += 1
            
            review_per_second = i / float(time.time() - start)

            sys.stdout.write('\rProcess:' + str(100*i/float(len(training_reviews)))[:4] + \
                             '% Speed(reviews/sec):' + str(review_per_second)[:5] + \
                            '% #Correct:' + str(correct_so_far) + ' #Trained:' + str(i+1) \
                            + ' Training Accuracy:' + str(correct_so_far*100/float(i+1))[:4])
                           
    def test(self, testing_reviews, testing_labels):
        correct = 0
        
        start = time.time()
        
        for i in range(len(testing_reviews)):
            pred = self.run(testing_reviews[i])
            if (pred == testing_labels[i]):
                correct += 1
                
            review_per_second = i / float(time.time() - start)
            
            sys.stdout.write('\rProcess:' + str(100*i/float(len(testing_reviews)))[:4] + \
                             '% Speed(reviews/sec):' + str(review_per_second)[:5] + \
                            '% #Correct:' + str(correct) + ' #Tested:' + str(i+1) \
                            + ' Testing Accuracy:' + str(correct*100/float(i+1))[:4])
            
        
    def run(self, review):
        # Input layer
              
        # Hidden layer
        self.hidden_layer *= 0
        unique_indices = set()
        for word in review.lower().split(' '):
            if word in self.word2index.keys():
                unique_indices.add(self.word2index[word])
        for idx in unique_indices:
            self.hidden_layer += self.weights_0_1[idx]
        
        # Output layer
        output = self.sigmoid(np.dot(self.hidden_layer, self.weights_1_2))
        
        if output[0] > 0.5:
            return 'POSITIVE'
        else:
            return 'NEGATIVE'
                
```
